```python
import numpy as np
import matplotlib.pyplot as plt
import f90nml
from matplotlib.colors import LogNorm
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
parameters = f90nml.read('box_parameters.namelist')
file_name = parameters['output']['output_file']
L_x, L_y = parameters['box']['length']
rho_0 = parameters['charge_distribution']['rho_zero']
x_0, y_0 = parameters['charge_distribution']['center']
R_x, R_y = parameters['charge_distribution']['width']
N = parameters['sampling']['n_max']
```

```python
solution = np.loadtxt(file_name, skiprows = 1, unpack = True)
x = solution[0]
y = solution[1]
V = solution[2:N + 2]
V_0 = 1.075 * abs(np.min(V))
V_gauge = V + V_0
onesArray = np.ones(N, float)
X = np.outer(x, onesArray)
Y = np.outer(onesArray, y)
print(len(X), len(Y))
conductor = np.zeros((N, N), float)
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
plt.imshow(V_gauge, norm = LogNorm(), extent =[x[0], x[N - 1], y[0], y[N - 1]])
plt.xlabel(r'$L_{x}$', fontsize = 14.5)
plt.ylabel(r'$L_{y}$', fontsize = 14.5)
plt.title('Potential inside Box', y = 1.025, fontsize = 20)
plt.show()
```

```python
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X, Y, conductor, color = '#7f7f7f')
ax.plot_surface(X, Y, V, cmap = 'viridis', alpha = 0.85)
displayInterval = 1.1 * np.max([L_x, L_y])
plt.xlim(0, displayInterval)
plt.ylim(0, displayInterval)
plt.xlabel(r'$x$', fontsize = 14.5)
plt.ylabel(r'$y$', fontsize = 14.5)
plt.show()
```
