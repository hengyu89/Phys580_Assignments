```python
# Importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
file_name = 'infinite_well_wf.dat'

all_data = np.loadtxt(file_name,skiprows=1)

all_data = all_data.transpose()

position_x = all_data[0]
ground_state_infinitewell = all_data[1]
first_excited_infinitewell = all_data[2]
second_excited_infinitewell = all_data[3]
```

```python
file_name2 = 'harmonic_oscillator_wf.dat'

all_data2 = np.loadtxt(file_name2,skiprows=1)

all_data2 = all_data2.transpose()

ground_state_harmonic = all_data2[1]
first_excited_harmonic = all_data2[2]
second_excited_harmonic = all_data2[3]
```

```python
file_name3 = 'woods_saxon_wf.dat'

all_data3 = np.loadtxt(file_name3,skiprows=1)

all_data3 = all_data3.transpose()

ground_state_woodssaxon = all_data3[1]
first_excited_woodssaxon = all_data3[2]
second_excited_woodssaxon = all_data3[3]
```

```python
file_name4 = 'woods_saxon_ener.dat'

all_data4 = np.loadtxt(file_name4,skiprows=1)

all_data4 = all_data4.transpose()

radius = all_data4[0]
first_lowest_energies = all_data[1]
second_lowest_energies = all_data[2]
third_lowest_energies = all_data[3]
```

```python
plt.plot(position_x, ground_state_infinitewell, label = 'infinite well')
plt.plot(position_x, ground_state_harmonic, label = 'harmonic oscillator')
plt.plot(position_x, ground_state_woodssaxon, label = 'woods saxon')

plt.xlabel('position of xi (fm)')
plt.ylabel('Ground State - Probability Density')

plt.legend()
plt.show()
```

```python
plt.plot(position_x, first_excited_infinitewell, label = 'infinite well')
plt.plot(position_x, first_excited_harmonic, label = 'harmonic oscillator')
plt.plot(position_x, first_excited_woodssaxon, label = 'woods saxon')

plt.xlabel('position of xi (fm)')
plt.ylabel('First Excited - Probability Density')

plt.legend()
plt.show()
```

```python
plt.plot(position_x, second_excited_infinitewell, label = 'infinite well')
plt.plot(position_x, second_excited_harmonic, label = 'harmonic oscillator')
plt.plot(position_x, second_excited_woodssaxon, label = 'woods saxon')

plt.xlabel('position of xi (fm)')
plt.ylabel('Third Excited - Probability Density')

plt.legend()
plt.show()
```

```python
plt.plot(radius, first_lowest_energies, label = 'Lowest Energy')
plt.plot(radius, second_lowest_energies, label = 'Second Lowest Energy')
plt.plot(radius, third_lowest_energies, label = 'Third Lowest Energy')

plt.xlabel('Radius (fm)')
plt.ylabel('Woods-Saxon (MeV)')

plt.legend()
plt.show()
```
